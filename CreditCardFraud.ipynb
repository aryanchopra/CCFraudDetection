```python
import pandas as pd
import numpy as np
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,accuracy_score,precision_score, recall_score,f1_score,classification_report
import seaborn as sns
from imblearn.over_sampling import SMOTE
from catboost import Pool, CatBoostClassifier, cv, CatBoostRegressor
```

```python
df=pd.read_csv('D:/Datasets/creditcard.csv')
```

```python
df.head()
```

```python
df['Class'].value_counts()
```

```python
df=df.drop(['Time','Amount'],axis=1)
```

```python
df.head()
```

```python
labels = df['Class']
```

```python
df=df.drop('Class',axis=1)
```

```python
df.isnull().sum()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df,labels,test_size=0.33,random_state=42)
```

```python
print(sum(ytrain==1),sum(ytrain==0))
```

```python
sm=SMOTE(random_state=42)
xtrainOS,ytrainOS=sm.fit_sample(xtrain,ytrain)
```

```python
print(sum(ytrainOS==1),sum(ytrainOS==0))
```

```python
model=CatBoostClassifier(iterations=100,depth=10,eval_metric='AUC',od_type='Iter',metric_period=1,random_seed=2020,od_wait=100)
```

```python
model.fit(xtrainOS,ytrainOS)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pred=model.predict(xtest)
```

```python
confusion_matrix(ytest,pred)
```

```python
acc=accuracy_score(ytest,pred)
print('Accuracy =' ,acc)
precision = precision_score(ytest,pred)
print('Precision =' ,precision)
auc_score=roc_auc_score(ytest,pred)
print('AUC =' ,auc_score)
recall = recall_score(ytest,pred)
print("Recall : ",recall )
f1score = f1_score(ytest,pred, average='macro')
print("F1 Score : ",f1score )
```
